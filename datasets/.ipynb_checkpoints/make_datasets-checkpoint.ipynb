```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option('precision', 16)
```

```python
import bz2
import pickle
import _pickle as cPickle

# Pickle a file and then compress it into a file with extension 
def compressed_pickle(title, data):
    with bz2.BZ2File(title + '.pbz2', 'w') as f: 
        cPickle.dump(data, f)

# Load any compressed pickle file
def decompress_pickle(file):
    data = bz2.BZ2File(file, 'rb')
    data = cPickle.load(data)
    return data
```

```python
dataframe = pd.read_csv('TRAINING_24March2022_UNNORMED.csv')
dataframe
```

```python
compressed_pickle('training_set_08May2022', dataframe)
```

```python
columns_to_norm = ['Theta', 'Pitch', 'Duty Cycle', 'Fill Factor', 'Mode', 'Lambda']
dataframe[columns_to_norm] = dataframe[columns_to_norm].apply(lambda x: (x-x.min()) / (x.max() - x.min()))
dataframe
```

```python
compressed_pickle('training_set_normalized_features_08May2022', dataframe)
```

```python
compressed_pickle('training_set_normalized_features', training_set)
```
